```python
#Task 1: Load and Explore Dataset
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (adjust path to where your dataset is saved)
try:
    df = pd.read_csv(r"C:\Users\SAIFUL\OneDrive\Desktop\Lizzie\PYTHON MODULE\archive.zip")
    print("✅ Dataset loaded successfully!")
except FileNotFoundError:
    print("❌ File not found. Please check the path again.")

# Display first few rows
df.head()

# Check structure
df.info()

# Check for missing values
df.isnull().sum()

# Clean dataset (if missing values exist)
df = df.dropna()   # drop missing rows)
print("✅ Missing values handled.")

#Task 2: Basic Data Analysis
# Basic statistics
df.describe()

# Grouping: Example - compute mean of BMI grouped by diabetes status (0 = No, 1 = Yes)
grouped_bmi = df.groupby("Diabetes_binary")["BMI"].mean()
print(grouped_bmi)

# Example grouping: Average age (GenHlth column) by diabetes status
grouped_health = df.groupby("Diabetes_binary")["GenHlth"].mean()
print(grouped_health)

# Task 3: Data Visualization
# Line chart: Average BMI across age groups
age_bmi = df.groupby("Age")["BMI"].mean()

plt.figure(figsize=(10,5))
plt.plot(age_bmi.index, age_bmi.values, marker="o")
plt.title("Average BMI Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Average BMI")
plt.grid(True)
plt.show()

# Bar chart: Average BMI by Diabetes status
plt.figure(figsize=(6,4))
sns.barplot(x="Diabetes_binary", y="BMI", data=df, errorbar=None,  hue="Diabetes_binary", palette="viridis", legend=False)

plt.title("Average BMI by Diabetes Status")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Average BMI")
plt.show()

# Histogram: Distribution of BMI
plt.figure(figsize=(8,5))
plt.hist(df["BMI"], bins=30, color="skyblue", edgecolor="black")
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

# Scatter plot: BMI vs. Age
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["BMI"], alpha=0.3, c=df["Diabetes_binary"], cmap="coolwarm")
plt.title("BMI vs. Age (Colored by Diabetes)")
plt.xlabel("Age Group")
plt.ylabel("BMI")
plt.colorbar(label="Diabetes (0=No, 1=Yes)")
plt.show()


```
